```python
import pandas as pd
```

```python
df = pd.read_csv('movie_data.csv')
```

```python
# Splitting the 'info' column into three separate columns
df[['Year', 'Genre', 'Interval']] = df['info'].str.split(', ', expand=True)
```

```python
df.drop('info',axis=1,inplace=True)
```

```python
# Renaming the 'title' column to 'Title'
df = df.rename(columns={'title': 'Title'})
```

```python
df['Genre'] = df['Genre'].str.replace(' & ','/')
```

```python
# Extracting genres and creating separate columns
df_genres = df['Genre'].str.get_dummies(sep='/')
```

```python
df_genres.head()
```

```python
# Combining the original dataframe with the genres dataframe
df = pd.concat([df, df_genres], axis=1)
```

```python
df.drop('Genre',axis=1,inplace=True)
```

```python
# Converting 'Year' column to int32 data type
df['Year'] = df['Year'].astype('int64')
```

```python
# Converting 'Year' column to int32 data type
df['Interval'] = df['Interval'].astype('str')
```

```python
import re
# Function to convert duration string to seconds
def convert_to_seconds(duration):
    hours = 0
    minutes = 0
    
    # Extract hours and minutes from the duration string using regular expressions
    hour_match = re.search(r"(\d+)h", duration)
    minute_match = re.search(r"(\d+)m", duration)
    
    # Check if hour and/or minute match is found
    if hour_match:
        hours = int(hour_match.group(1))
    if minute_match:
        minutes = int(minute_match.group(1))
    
    # Calculate total duration in seconds
    total_seconds = hours * 3600 + minutes * 60
    return total_seconds
```

```python
# Convert the 'Interval' column to seconds
df['Interval'] = df['Interval'].apply(convert_to_seconds)
```

```python
# Converting 'Year' column to int32 data type
df['Title'] = df['Title'].astype('str')
```
